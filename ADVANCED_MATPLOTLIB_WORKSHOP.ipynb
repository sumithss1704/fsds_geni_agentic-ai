```python
import pandas as pd
```

```python
data={
    "Month":['Jan','Feb','Mar','Apr','May','Jun'],
    "Sale" :[10000,12000,15000,13000,17000,16000],
    "Profit":[2000,3000,4000,2500,3500,3000]
}
df=pd.DataFrame(data)
print(df)
```

```python
df[['Month','Sale']]
```

#1 - line plot monthly sales over

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,3))
plt.plot(df['Month'],df['Sale'],color='blue',marker='o',linestyle='-',label='Sales')
plt.title('Sales Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.show()
```

#2-BARPLOT MONTH VS PROFIT


```python
plt.figure(figsize=(6,3))
width=0.3
plt.bar(df['Month'],df['Sale'],width=width,color='skyblue',label='Sales')
plt.bar(df['Month'],df['Profit'],width=width,color='green',label='Profit',bottom=df['Sale'])
plt.title('Sales and Profit Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Amount($)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# 3-PIE CHART PROFIT VS MONTH

```python
from enum import auto
plt.figure(figsize=(10,5))
plt.pie(df['Profit'],labels=df['Month'],autopct='%1.2f%%',startangle=140,colors=plt.cm.Paired.colors)
plt.title('Profit Distribution Over Months')
plt.axis('equal')
plt.show()
```

#4-Scatter plot


```python
plt.figure(figsize=(10,5))
plt.scatter(df['Sale'],df['Profit'],color='green',s=100,edgecolors='black')
plt.title('Sales vs Profit Scatter Plot')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#5-HISTOGRAM

```python
plt.figure(figsize=(8,4))
plt.hist(df['Sale'],bins=5,color='red',edgecolor='black')
plt.title('Sales Distribution Histogram')
plt.xlabel('Sales')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#BOXPLOT

```python
plt.figure(figsize=(8,4))
plt.boxplot(df['Profit'])
plt.title('Sales Box Plot')
plt.ylabel('Sales')
plt.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import gradio as gr

data={
    "Month":['Jan','Feb','Mar','Apr','May','Jun'],
    "Sale" :[10000,12000,15000,13000,17000,16000],
    "Profit":[2000,3000,4000,2500,3500,3000]
}
df=pd.DataFrame(data)
def generate_plot(plot_type):
    fig=plt.figure(figsize=(10,5))

    if plot_type=='Line Plot':
        plt.plot(df['Month'],df['Sale'],color='blue',marker='o',linestyle='-',label='Sales')
        plt.title('Sales Trends Over Months')
        plt.xlabel('Month')
        plt.ylabel('Sales')
        plt.grid(True)
        plt.legend()
    elif plot_type=='stacked Bar Plot':
      fig.set_size_inches(10,5)
      width=0.3
      plt.bar(df['Month'],df['Sale'],width=width,color='skyblue',label='Sales')
      plt.bar(df['Month'],df['Profit'],width=width,color='green',label='Profit',bottom=df['Sale'])
      plt.title('Sales and Profit Trends Over Months')
      plt.xlabel('Month')
      plt.ylabel('Amount($)')
      plt.grid(True)
      plt.legend()

    elif plot_type=='Pie Chart':
      fig.set_size_inches(10,5)
      plt.pie(df['Profit'],labels=df['Month'],autopct='%1.2f%%',startangle=140,colors=plt.cm.Paired.colors)
      plt.title('Profit Distribution Over Months')

    elif plot_type=='Scatter Plot':
      plt.scatter(df['Sale'],df['Profit'],color='green',s=100,edgecolors='black')
      plt.title('Sales vs Profit Scatter Plot')
      plt.xlabel('Sales')
      plt.ylabel('Profit')
      plt.grid(True)

    elif plot_type=='Histogram':
      plt.hist(df['Sale'],bins=5,color='red',edgecolor='black')
      plt.title('Sales Distribution')
      plt.xlabel('Sales')
      plt.ylabel('Frequency')

    elif plot_type=='Box Plot':
      plt.boxplot(df['Profit'],vert=False,patch_artist=True,boxprops=dict(facecolor='skyblue'))
      plt.title('Sales Box Plot')
      plt.ylabel('Sales')
    plt.tight_layout()
    return fig


demo=gr.Interface(
    fn=generate_plot,
    inputs=gr.Radio(['Line Plot','stacked Bar Plot','Pie Chart','Scatter Plot','Histogram','Box Plot'],label="choose plot type"),
    outputs=gr.Plot(label="Sales data visualization"),
    title='Sales and profit visual insight',
    description='choose the type to visualize the data'
)

demo.launch()



```
